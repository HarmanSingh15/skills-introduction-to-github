### Reorgainizing Array

```python
before = np.array([[1,2,3,4], [5,6,7,8]])
print(before)

try:
  after = before.reshape((2,3))
  print(after)
except Exception as e:
  print(e)

after = before.reshape((4,2))
print(after)
```

```python
before.flatten()
```

```python
after.reshape((2,2,2))
```

```python
after = before.reshape((4,-1))
print(after)

print()
after = before.reshape((-1,2))
print(after)
```
