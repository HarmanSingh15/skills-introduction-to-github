```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


class Knnclassifier:
  def __init__(self, k=5):
    self.k = k

  def fit(self,x,y):
    self.x = x
    self.y = y


  def distance_formula(self, p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))


  def predict_1(self, x_test):
    distances = []
    for i in range(len(self.x)):
      distances.append((self.distance_formula(x_test, self.x[i]),self.y[i]))
    distances = sorted(distances, key=lambda x:x[0])[:self.k]

    klables=[]
    for i in range (self.k):
      klables.append(distances[i][1])

    labels, count = np.unique(klables, return_counts=True)
    predLables = labels[np.argmax(count)]
    prob = np.max(count)/self.k
    print("probablity:", prob)
    return predLables


  def predict(self, x_test):
    predictions = []
    for i in range(len(x_test)):
      predictions.append(self.predict_1(x_test[i]))
    return predictions
```
