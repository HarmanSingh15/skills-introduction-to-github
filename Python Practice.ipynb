```python
#Write a python program to compute multiplication of two given matrices
def multiply(a, b):
  m1 = len(a)
  n1 = len(a[0])
  m2 = len(b)
  n2 = len(b[0])

  if n1 != m2:
    print("Not possible")
    return -1

  res = []
  for i in range(m1):
    new_row = [0] * n2
    res.append(new_row)

  for i in range(m1):
    for j in range(n2):
      dot_product = 0
      for k in range(n1):
        dot_product += a[i][k]*b[k][j]

      res[i][j] = dot_product

  return res
```

```python
# Write a program to compute the determinant of a given square matrix (n <= 3)
```

```python
def determinant(matrix):
  n = len(matrix)

  for row in matrix:
    if len(row) != n:
      print("Matrix is not square")
      return -1

  if n == 1:
    return matrix[0][0]

  elif n == 2:
    a, b = matrix[0]
    c, d = matrix[1]
    return a*d - b*c

  else:
    a, b, c = matrix[0]
    d, e, f = matrix[1]
    g, h, i = matrix[2]
    return (a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g))
```

```python
matrix1 = [[5]]
matrix2 = [[1,2], [3,4]]
matrix3 = [[1,2,3],[4,5,6],[7,8,9]]

print(determinant(matrix1))
print(determinant(matrix2))
print(determinant(matrix3))
```

```python
#Write a python program to find vector norm of a vector

def norm(x, n):
  sum = 0
  for num in x:
    sum += abs(num)**n

  norm = sum**(1/n)
  return norm

vector = [3,4]
n = 2

print(norm(vector, n))
```

```python
# Write a program to compute the trace of a given matrix
def trace(matrix):
  n = len(matrix)

  for row in matrix:
    if len(row) != n:
      print("Matrix is not square")
      return -1

  sum = 0
  for i in range(n):
    sum += matrix[i][i]

  return sum
```

```python
matrix1 = [[1,2], [3,4]]
matrix2 = [[1,0,0], [0,2,0], [0,0,3]]
print(trace(matrix1))
print(trace(matrix2))
```

```python
#Write a python program to find the solutions of 2 equations in 2 variable
#example: Eq1 - 2x+y=3 and Eq2 - 3x-y=1

def solve_eq(a1,b1,c1,a2,b2,c2):
  determinant = a1*b2 - a2*b1

  if determinant == 0:
    print("No solution")
    return -1

  x = (c1*b2 - c2*b1)/determinant
  y = (c2*a1 - c1*a2)/determinant

  return x,y

a1,b1,c1 = 2, 1, 3
a2,b2,c2 = 3,-1, 1

x,y = solve_eq(a1,b1,c1,a2,b2,c2)
print(x, y)
```

```python
#Calculate the transpose of a matrix
def transpose(a):
  res = []
  for i in range(len(a[0])):
    new_row = [0] * len(a)
    res.append(new_row)

  for i in range(len(a)):
    for j in range(len(a[0])):
      res[j][i] = a[i][j]

  return res
```

```python
A = [[1,2,3], [4,5,6]]
print(transpose(A))
```

```python
# Program to check whether a matrix is orthogonal or not
def check_identity_matrix(a):
  row = len(a)
  col = len(a[0])

  if row != col:
    return False

  for i in range(row):
    for j in range(col):
      if i == j and a[i][j] != 1:
        return False
      if i != j and a[i][j] != 0:
        return False

  return True
```

```python
def check_orthogonal(a):
  a_transpose = transpose(a)
  product = multiply(a, a_transpose)
  return check_identity_matrix(product)
```

```python
a = [[1,0], [0,-1]]
print(check_orthogonal(a))
```

```python
# Check if a matrix is symmetric or not
def symmetric(a):
  for i in range(len(a)):
    for j in range(len(a[0])):
      if(a[j][i] != a[i][j]):
        return False

  return True
```

```python
a = [[1,3,5], [3,2,4], [5,4,1]]
print(symmetric(a))
```
