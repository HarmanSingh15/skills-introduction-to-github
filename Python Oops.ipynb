```python
# Write a program to create a shopping cart. Include methods for adding and removing items and
# calculate total price. For shopping cart you can take attributes as item name and cost
class ShoppingCart:
  def __init__(self):
    self.catalog = {}

  def add(self, item_name, cost):
    self.catalog[item_name] = cost

  def remove(self, item_name):
    for key in self.catalog.keys():
      if key == item_name:
        self.catalog[key] = 0
        break

  def total_price(self):
    sum = 0
    for val in self.catalog.values():
      sum += val
    return sum

obj = ShoppingCart()
obj.add("toy_car", 100)
obj.add("apple", 10)
obj.add("ball", 50)

print(obj.total_price())

obj.remove("apple")

print(obj.total_price())
```

```python
class calculator:
  def add(self, x, y):
    return x + y

  def sub(self, x, y):
    return x - y

  def mul(self, x, y):
    return x * y

  def div(self, x, y):
    if y != 0:
      return x/y
    return -1

  def power(self, x, y):
    return x**y

obj = calculator()
print(obj.add(5,11))
print(obj.sub(4,2))
print(obj.mul(5, 3))
print(obj.div(5, 0))
print(obj.div(5, 5))
print(obj.power(5, 2))
```

### Inheritance

```python
class Person(object):
  def __init__(self, name, id):
    self.name = name
    self.id = id

  def display(self):
    print(self.name)
    print(self.id)

  def details(self):
    print(f"My name is {self.name}")
    print(f"My id is {self.id}")


class student(Person):
  def __init__(self, name, id, rollno, marks):
    self.rollno = rollno
    self.marks = marks
    Person.__init__(self, name, id)

  def details(self):
    print(f"My name is {self.name}")
    print(f"My id is {self.id}")
    print(f"My rollno is {self.rollno}")
    print(f"My marks is {self.marks}")
```

```python
obj = student("Rahul", 123, 2020201, 450)
obj.display()
obj.details()
```

```python
class bird:
  def abc(self):
    print("hello")

  def fly(self):
    print("Birds can fly")

class crow(bird):
  def fly(self):
    print("crow can fly")

class ostrich(bird):
  def fly(self):
    print("ostrich cannot fly")
```

```python
obj_bird = bird()
obj_crow = crow()
obj_ostrich = ostrich()

obj_bird.abc()
obj_bird.fly()

obj_crow.abc()
obj_crow.fly()

obj_ostrich.abc()
obj_ostrich.fly()
```

```python
class person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def display(self):
    print(self.name, self.age)

class person_child(person):
  def __init__(self, name, age):
    self.new_name = name
    self.new_age = age
    super().__init__("Rahul", 12)

  def display2(self):
    print(self.new_name, self.new_age)

obj = person_child("abc", 56)
obj.display()
obj.display2()
```

```python
# Multiple Inheritance
class parent1:
  def __init__(self):
    self.s1 = "abc"
    print("parent1")

class parent2:
  def __init__(self):
    self.s2  = "xyz"
    print("parent2")

class child(parent1, parent2):
  def __init__(self):
    parent1.__init__(self)
    parent2.__init__(self)
    print("child")

  def print(self):
    print(self.s1 + self.s2)

obj = child()
obj.print()
```

```python
#Multilevel Inheritance
class parent:
  def __init__(self, name):
    self.name = name

  def get_name(self):
    return self.name

class child(parent):
  def __init__(self, name, age):
    parent.__init__(self, name)
    self.age = age

  def get_age(self):
    return self.age

class grand_child(child):
  def __init__(self, name, age, number):
    child.__init__(self, name, age)
    self.number = number

  def get_number(self):
    return self.number

obj = grand_child("abc", 18, 8388233203)
print(obj.get_name(), obj.get_age(), obj.get_number())
```

```python
def temp(a, b, *args):
  sum = a + b
  for i in args:
    sum += i

  return sum

print(temp(4,5,6,7,8,9,9))
```

```python
def temp(a, b, *args, **kwargs):
  sum = a + b
  for i in args:
    sum += i

  for key, value in kwargs.items():
    print(key, value)
  return sum

print(temp(4,5,6,7,8,9,9, name = "abc", number = 123, temp = [1,2,3]))
```

```python
def new_intro(obj):
  return obj.intro()

class dog:
  def intro(self):
    return "bark"

class cat:
  def intro(self):
    return "meow"

obj1 = dog()
obj2 = cat()

print(new_intro(obj1))
print(new_intro(obj2))
```

```python
from abc import abstractmethod

class animal:
  @abstractmethod
  def intro(self):
    pass

class dog(animal):
  def intro(self):
    return "bark"

class cat(animal):
  def intro(self):
    return "meow"

obj1 = dog()
obj2 = cat()

print(obj1.intro())
print(obj2.intro())
```

```python
class parent:
  def __init__(self):
    self._a = 2

class child(parent):
  def __init__(self):
    parent.__init__(self)
    print("protected memeber of parent", self._a)

    self._a = 5
    print("protected memeber of parent again", self._a)

obj1 = child()
obj2 = parent()

print(obj1._a)
print(obj2._a)
```
